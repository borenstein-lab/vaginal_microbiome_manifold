#### PCA for cros data

```python
def get_pca_reg(df): 
    pca = PCA(n_components = 50)
    pca.fit(df)
    pca_array = pca.transform(df)
    pca_df = pd.DataFrame(data = pca_array, columns = pca_cols, index = df.index)
    
    loadings_df = pd.DataFrame(pca.components_.T, columns = pca_cols, index = df.columns)
   
    return pca_df, loadings_df
```

```python
ps_pca_df, ps_loadings = get_pca_reg(int_ps_df)
```

#### Project PCA on temp data

```python
def get_project(loadings, df):
    new_pca_df = pd.DataFrame(index = df.index)
    loadings_arr = np.array(ps_loadings)
    
    ## Center df
    mean_features = df.mean(axis = 0)
    df_mean_features = df - mean_features
    
    ## Matmul
    for i in range(0, len(loadings_arr.T)):
        res_arr = np.matmul(loadings_arr[:, i], np.array(df_mean_features.T))
        new_pca_df[loadings.columns[i]] = res_arr
        
    return new_pca_df
```

```python
def get_paga(andata, n_neighbors, n_pcs, root = '0'):
    sc.pp.neighbors(andata, n_neighbors = n_neighbors, n_pcs = n_pcs, metric = 'euclidean')
    sc.tl.draw_graph(andata)
    sc.pl.draw_graph(andata, color = 'community', legend_loc = 'on data')
    
    sc.tl.leiden(andata, resolution = 1.0)
    sc.tl.paga(andata, groups = 'leiden')
    
    print(andata.obs['leiden'].cat.categories)
    andata.obs['leiden_anno'] = andata.obs['leiden']
    sc.tl.paga(andata, groups = 'leiden_anno')
    sc.pl.paga(andata, threshold = 0.03, show = False)
    
    andata.uns['iroot'] = np.flatnonzero(andata.obs['leiden_anno']  == root)[0] ## BV
    sc.tl.dpt(andata)
    sc.pl.draw_graph(andata, color = ['leiden_anno', 'community', 'dpt_pseudotime'], legend_loc = 'on data')
    
    return
```

#### Get the nearest neighbor of temp data

```python
def get_pcoa(df, metric):
    arr = df.to_numpy()
    bc_df = beta_diversity(counts = arr, ids = list(df.index), metric = metric, validate = False)

    return bc_df.to_data_frame()
```

```python
def get_filt_distance_df(distance_df, ps_meta_df, meta_df):
    cols_to_keep = meta_df.index
    rows_to_keep = ps_meta_df.index
    
    filt_distance_df = distance_df[cols_to_keep]
    filt_distance_df = filt_distance_df[filt_distance_df.index.isin(rows_to_keep)]
    
    return filt_distance_df
```

```python
def get_knn(filt_distance_df, meta_df):
    res_df = pd.DataFrame(columns = ['closest_sample', 'distance'], index = meta_df.index)
    for col in filt_distance_df.columns:
        min_distance = filt_distance_df[col].min()
        min_indx = filt_distance_df[col].idxmin()
        closest_sample = min_indx
                
        ## insert results to res_df
        res_df.loc[col, 'distance'] = min_distance
        res_df.loc[col, 'closest_sample'] = closest_sample
        
    return res_df
```

```python
distance_df = get_pcoa(both_df, 'braycurtis')
filt_distance_df = get_filt_distance_df(distance_df, ps_meta_df, temp_meta_df)
knn_res_df = get_knn(filt_distance_df, temp_meta_df)
```

```python
def get_scatter_plot(df, tit, col, sym, three_d = False):    
    pio.templates.default = 'none'
    
    ## Plots    
    if three_d:
        fig = px.scatter_3d(df, x = df.columns[0], y = df.columns[1], z = df.columns[3], 
                        color = col, symbol = sym, 
                        opacity = 0.5, width = 800, height = 800,
                        title = tit,
                        color_discrete_sequence = px.colors.sequential.Plasma_r)
        symbols_dict = symbols_3d
    else:
        fig = px.scatter(df, x = df.columns[0], y = df.columns[1],  
                        color = col, symbol = sym, 
                        opacity = 0.5, width = 800, height = 800,
                        title = tit,
                        color_discrete_sequence = px.colors.sequential.Plasma_r)
        symbols_dict = symbols

    fig.update_layout(showlegend = True)
    fig.update_layout(coloraxis_colorbar = dict(yanchor = "top", y = 1, x = -0.35,
                                          ticks = "outside"))       
    fig.update_traces(marker = dict(size = 13,
                              line = dict(width = 0.8,
                                        color = 'black')))
#     set all symbols in fig
    if sym is 'community':
        if col is 'mt_pseudotime':
            for i, d in enumerate(fig.data):
                fig.data[i].marker.symbol = symbols_dict[fig.data[i].name]
    
    fig.show()
    
    return fig
```

```python
def get_paired_plt(df, legend_loc):
    ## Variables
    figsize_x = 25
    figsize_y = 8
    ax_size = 15
    lab_size = 18
    point_size = 130
    leg_size = 12
    x_col_name = df.columns[0]
    y_col_name = df.columns[1]
    
    ## Create coordinates for lines
    X_coords = np.array([df.loc[df['DB_type'] == 'cros', x_col_name], 
                         df.loc[df['DB_type'] == 'temp', x_col_name]])
    Y_coords = np.array([df.loc[df['DB_type'] == 'cros', y_col_name], 
                         df.loc[df['DB_type'] == 'temp', y_col_name]])
    
    ##
    figure, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey = True, figsize = (figsize_x, figsize_y))
    shapes = {'cros': 's', 'temp': 'o'}
    
    ## Plot 1- DB_type
    ax1.plot(X_coords, Y_coords, color = 'dimgrey', zorder = -1, linewidth = 0.5)
    colors = {'cros': 'firebrick', 'temp': 'teal'}
    ax1.scatter(df.loc[df['DB_type'] == 'cros', x_col_name], 
                df.loc[df['DB_type'] == 'cros', y_col_name], 
                s = point_size, edgecolor = 'black', c = 'firebrick', alpha = 0.6, marker = 's', label = 'cros-sectional')
    ax1.scatter(df.loc[df['DB_type'] == 'temp', x_col_name], 
                df.loc[df['DB_type'] == 'temp', y_col_name], 
                s = point_size, edgecolor = 'black', c = 'teal', alpha = 0.6, marker = 'o', label = 'temporal')
    ax1.set_xlabel(x_col_name, size = lab_size)
    ax1.set_ylabel(y_col_name, size = lab_size)
    ax1.tick_params(axis = 'both', which = 'major', labelsize = ax_size)
    ax1.legend(loc = legend_loc, fontsize = leg_size)
    
    ## Plot 2- comminty
    ax2.plot(X_coords, Y_coords, color = 'dimgrey', zorder = -1, linewidth = 0.5)
    community_colors = {'I':'crimson', 'II':'orange', 'III':'chocolate', 'V':'gold', 
              'IV':'steelblue', 'IV-A':'aqua', 'IV-B':'springgreen'}
    plt2 = ax2.scatter(df.loc[df['DB_type'] == 'cros', x_col_name], 
            df.loc[df['DB_type'] == 'cros', y_col_name], 
            s = point_size, edgecolor = 'black', c = df.loc[df['DB_type'] == 'cros', 'community'].map(community_colors), 
            alpha = 0.6, marker = 's', label = 'cros- sectional')
    ax2.scatter(df.loc[df['DB_type'] == 'temp', x_col_name], 
            df.loc[df['DB_type'] == 'temp', y_col_name], 
            s = point_size, edgecolor = 'black', c = df.loc[df['DB_type'] == 'cros', 'community'].map(community_colors), 
            alpha = 0.6, marker = 'o', label = 'temporal')
    ax2.set_xlabel(x_col_name, size = lab_size)
    ax2.tick_params(axis = 'both', which = 'major', labelsize = ax_size)
    ax2.legend(loc = legend_loc, fontsize = leg_size)

    
    ## Plot 3- mt_pseudotime
    ax3.plot(X_coords, Y_coords, color = 'dimgrey', zorder = -1, linewidth = 0.5)
    pl3 = ax3.scatter(df.loc[df['DB_type'] == 'cros', x_col_name], 
            df.loc[df['DB_type'] == 'cros', y_col_name], 
            s = point_size, edgecolor = 'black', c = df.loc[df['DB_type'] == 'cros', 'mt_pseudotime'], 
            alpha = 0.6, marker = 's', label = 'cros- sectional')
    ax3.scatter(df.loc[df['DB_type'] == 'temp', x_col_name], 
            df.loc[df['DB_type'] == 'temp', y_col_name], 
            s = point_size, edgecolor = 'black', c = df.loc[df['DB_type'] == 'temp', 'mt_pseudotime'], 
            alpha = 0.6, marker = 'o', label = 'temporal')
    ax3.set_xlabel(x_col_name, size = lab_size)
    ax3.tick_params(axis = 'both', which = 'major', labelsize = ax_size)
    ax3.legend(loc = legend_loc, fontsize = leg_size)
    cbar = figure.colorbar(pl3)
    cbar.set_label("Pseudotime", labelpad =+ 0.5)
    
    return
```
