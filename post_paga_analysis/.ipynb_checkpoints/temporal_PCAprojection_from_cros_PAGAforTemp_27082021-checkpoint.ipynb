```python
def get_append(first_df, second_df, filler):
    if second_df is not ps_meta_df:
        both_df = pd.concat([first_df, second_df], join = 'inner')
    else:
        both_df = first_df.append(second_df, sort = False)
        
    both_df_nona = both_df.fillna(filler)
    
    return both_df_nona
```

#### PCA for cros data

```python
def get_pca_reg(df): 
    pca = PCA(n_components = 50)
    pca.fit(df)
    pca_array = pca.transform(df)
    pca_df = pd.DataFrame(data = pca_array, columns = pca_cols, index = df.index)
    
    loadings_df = pd.DataFrame(pca.components_.T, columns = pca_cols, index = df.columns)
   
    return pca_df, loadings_df
```

```python
ps_pca_df, ps_loadings = get_pca_reg(int_ps_df)
```

#### Project PCA on temp data

```python
def get_project(loadings, df):
    new_pca_df = pd.DataFrame(index = df.index)
    loadings_arr = np.array(ps_loadings)
    
    ## Center df
    mean_features = df.mean(axis = 0)
    df_mean_features = df - mean_features
    
    ## Matmul
    for i in range(0, len(loadings_arr.T)):
        res_arr = np.matmul(loadings_arr[:, i], np.array(df_mean_features.T))
        new_pca_df[loadings.columns[i]] = res_arr
        
    return new_pca_df
```

```python
temp_pca_df = get_project(ps_loadings, int_temp_df)
display(temp_pca_df)
```

```python
def get_paga(andata, n_neighbors, n_pcs, root = '0'):
    sc.pp.neighbors(andata, n_neighbors = n_neighbors, n_pcs = n_pcs, metric = 'euclidean')
    sc.tl.draw_graph(andata)
    sc.pl.draw_graph(andata, color = 'community', legend_loc = 'on data')
    
    sc.tl.leiden(andata, resolution = 1.0)
    sc.tl.paga(andata, groups = 'leiden')
    
    print(andata.obs['leiden'].cat.categories)
    andata.obs['leiden_anno'] = andata.obs['leiden']
    sc.tl.paga(andata, groups = 'leiden_anno')
    sc.pl.paga(andata, threshold = 0.03, show = False)
    
    andata.uns['iroot'] = np.flatnonzero(andata.obs['leiden_anno']  == root)[0] ## BV
    sc.tl.dpt(andata)
    sc.pl.draw_graph(andata, color = ['leiden_anno', 'community', 'dpt_pseudotime'], legend_loc = 'on data')
    
    return
```

```python
def get_partial_pseudotime(df, leiden_lst, column_name):
    full_column_name = 'branch_' + column_name
    df[full_column_name] = [df['dpt_pseudotime'].iloc[i] if df['leiden_anno'].iloc[i] in leiden_lst 
                              else np.nan 
                              for i in range(0, len(df))]
    return df

## PCA
branch_df_pca = pd.DataFrame(adata_temp_pca.obs['leiden_anno'])
branch_df_pca['dpt_pseudotime'] = adata_temp_pca.obs['dpt_pseudotime']

branch_df_pca = get_partial_pseudotime(branch_df_pca, ['9', '12', '20', '11', '1', '19', '15', '7', '8', '0', '5'], 'I/IV') 
branch_df_pca = get_partial_pseudotime(branch_df_pca, ['24', '18', '19', '15', '7', '8', '0', '5'], 'II/IV')
branch_df_pca = get_partial_pseudotime(branch_df_pca, ['23', '4', '14', '16', '2', '13', '10', '6', '3', '0', '5'], 'III/IV')
branch_df_pca = get_partial_pseudotime(branch_df_pca, ['9', '12', '20', '11', '17', '6', '23', '4', '14', '16', '2', '13', '10'], 'I/III')

##
filter_col = [col for col in branch_df_pca if col.startswith('branch_')]
for col in branch_df_pca[filter_col].columns:
    new_name = 'mt_' + col
    branch_df_pca[new_name] = 1 - branch_df_pca[col]
    
branch_df_pca['mt_pseudotime'] = 1 - branch_df_pca['dpt_pseudotime']
branch_df_pca = pd.concat([branch_df_pca, temp_meta_df], axis = 1, join = 'inner')
display(branch_df_pca)
```

```python
def get_shannon(df):
    shannon_vec = alpha_diversity(counts = df.values, metric = 'shannon')
    shannon_df = pd.DataFrame(shannon_vec, columns = ['shannon_index'])
    shannon_df = shannon_df.set_index(df.index)
    
    return shannon_df
```

```python
def get_order(meta_df, shannon_df):
    filter_col = [col for col in meta_df if col.startswith('mt_branch_')]
    branch_df = meta_df[filter_col]
    
    final_df = pd.concat([shannon_df, branch_df], axis = 1).reindex(branch_df.index)
    
    ## log df
    shannon_log_df = pd.DataFrame(index = branch_df.index)
    shannon_log_df['shannon_index'] = np.log10(shannon_df['shannon_index'])
    final_log_df = pd.concat([shannon_log_df, branch_df], axis = 1).reindex(branch_df.index)
    
    return final_df, final_log_df
```

```python
def get_trend_plot(df, tit, color):
    arr_x = list(df.iloc[:, 1])
    arr_y =  list(df.iloc[:, 0])
    pio.templates.default = 'none'
    
    fig = px.scatter(x = arr_x, y = arr_y, 
                        color = df[color], 
                        labels = {
                        'x': "Pseudotime",
                        'y': "Shannon diversity index"},
                        opacity = 0.7, size_max = 15, width = 800, height = 800,
                        title = tit)
    
    # linear regression
    regline = sm.OLS(arr_y, sm.add_constant(arr_x)).fit()

    # add linear regression line for whole sample
    fig.add_traces(go.Scatter(x = arr_x, y = regline.fittedvalues,
                          mode = 'lines',
                          marker_color = 'black',
                          name = 'trend-line'))
    
    # add annotation
    fig.add_annotation(dict(font=dict(color = 'black',size = 16),
                                        x = 0,
                                        y = -0.12,
                                        showarrow = False,
                                        text = 'R squared = ' + str(round(regline.rsquared, 3)) + '\n' + 
                                               'P.value = ' + str(round(regline.pvalues[0], 6)),
                                        textangle = 0,
                                        xanchor = 'left'))
    fig.update_layout(font = dict(size = 19))
        
    return fig
```

```python
def get_reg(df, meta_df, color):
    fig_dict1 = {}
    fig_dict2 = {}
    spearman_index = df.columns[1:].str.replace('mt_branch_', '')
    spearman_df = pd.DataFrame(index = spearman_index, columns = ['rho', 'p.value'])
    
    for col in df.columns[1:]:
        new_df = df[['shannon_index', col]].copy()
        new_df[col] = new_df[col].astype(float)
        new_df["shannon_index"] = new_df["shannon_index"].astype(float)
        new_df = new_df[new_df[col].notna()]
        plot_df = pd.concat([new_df, meta_df[['community', 'leiden_anno']]], axis = 1).reindex(new_df.index)
        if 'IV/IV' not in col:
            plot_df = plot_df[~plot_df.community.str.contains('IV',case=False)]
        tit = col.replace('mt_branch_', '')
        
        ## Plot df with trendline
        fig1 = get_trend_plot(plot_df, tit, color)
        fig1.show()
        fig_dict1[col] = fig1
        
        ## Spearman correlation
        rho, p = spearmanr(plot_df.iloc[:, 0], plot_df.loc[:, col], nan_policy = 'omit')
        spearman_df.at[tit, 'rho'] = rho
        spearman_df.at[tit, 'p.value'] = p
        
    return fig_dict1, spearman_df
```

```python
def all_shannon_functions(df, meta_df, color):
    shannon_df = get_shannon(df)
    new_df, new_log_df = get_order(meta_df, shannon_df)
    fig_dict, spearman_df = get_reg(new_df, meta_df, color)
    fig_dict_log, spearman_df2 = get_reg(new_log_df, meta_df, color)
    
    return fig_dict, fig_dict_log, spearman_df
```

#### Save PCA temporal PAGA

```python
def save_results(andata, branch_df, df, fig_dict, spearman_df, save_file_path, method_file_path, file_name):
    params_df = pd.DataFrame({'n_neighbors': [andata.uns['neighbors']['params']['n_neighbors']],
                         'n_pcs': [andata.uns['neighbors']['params']['n_pcs']]})

    pcoa_df = pd.DataFrame(andata.obsm['X_pca'][:,:3])
    pcoa_df = pcoa_df.set_index(branch_df.index)
    pcoa_df.columns = ['PCA1', 'PCA2', 'PCA3']
    pcoa_df['dpt_pseudotime'] = andata.obs['dpt_pseudotime']
    pcoa_df['mt_pseudotime'] = 1 - pcoa_df['dpt_pseudotime']

    graph_df = pd.DataFrame(andata.obsm['X_draw_graph_fa'])
    graph_df.columns = ['fa1', 'fa2']
    graph_df = graph_df.set_index(branch_df.index)
    graph_df['dpt_pseudotime'] = andata.obs['dpt_pseudotime']
    graph_df['mt_pseudotime'] = 1 - pcoa_df['dpt_pseudotime']
    graph_df = pd.concat([graph_df, pd.DataFrame(andata.obsm['X_diffmap'], index = branch_df.index)], axis = 1)

    ## Save file
    path = save_file_path + method_file_path + file_name
    writer = pd.ExcelWriter(path, engine = 'xlsxwriter')
    branch_df.to_excel(writer, sheet_name = 'metadata_pseudo')
    pcoa_df.to_excel(writer, sheet_name = 'pcoa_pseudo')
    df.to_excel(writer, sheet_name = 'abundance_pseudo')
    graph_df.to_excel(writer, sheet_name = 'graphs_values_pseudo')
    params_df.to_excel(writer, sheet_name = 'params')
    spearman_df.to_excel(writer, sheet_name = 'spearman_shannon')
    writer.save()
    writer.close()
    
    ## Save figs
    png_path = save_file_path + method_file_path + 'Shannon_figures/'
    get_save_plots(fig_dict, png_path)
    
    return
```

```python
def get_save_plots(fig_dict, png_path):
    if not os.path.exists(png_path):
        os.mkdir(png_path)
    for key, value in fig_dict.items():
        new_key = key.replace("/", '_')
        pio.write_image(value, file = png_path + str(new_key) + '.png', 
                        format = "png",  
                        validate = True, engine = 'auto')
```
