#### PCA for cros data

```python
def get_pca_reg(df): 
    pca = PCA(n_components = 50)
    pca.fit(df)
    pca_array = pca.transform(df)
    pca_df = pd.DataFrame(data = pca_array, columns = pca_cols, index = df.index)
    
    loadings_df = pd.DataFrame(pca.components_.T, columns = pca_cols, index = df.columns)
   
    return pca_df, loadings_df
```

```python
ps_pca_df, ps_loadings = get_pca_reg(ps_df)
```

#### Order loadings by intersect species

```python
def get_loadings_int(both_df, loadings_df):
    int_spec_lst = both_df.columns
    both_loadings = loadings_df.loc[int_spec_lst, :]
    not_int_spec = pd.concat([both_loadings, loadings_df]).drop_duplicates(keep = False)
    
    zero_df = pd.DataFrame(columns = not_int_spec.columns, index = not_int_spec.index)
    zero_df.fillna(0, inplace = True)
    
    int_loadings = pd.concat([both_loadings, zero_df])
    int_loadings.reindex(loadings_df.index)
    
    return int_loadings, both_loadings
```

```python
int_loadings, both_loadings = get_loadings_int(both_df, ps_loadings)
display(both_loadings)
```

#### Project PCA on temp data

```python
def get_project(loadings, proj_df):
    new_pca_df = pd.DataFrame(index = proj_df.index)
    loadings_arr = np.array(loadings)
    
    ## Center df
    mean_features = proj_df.mean(axis = 0)
    df_mean_features = proj_df - mean_features
    
    ## Matmul
    for i in range(0, len(loadings_arr.T)):
        res_arr = np.matmul(loadings_arr[:, i], np.array(df_mean_features.T))
        new_pca_df[loadings.columns[i]] = res_arr
        
    return new_pca_df
```

#### Find euclidean nearest neighbor

```python
def get_euc_nn(ps_df, temp_df):
    dist_arr = euclidean_distances(ps_df, temp_df)
    dist_df = pd.DataFrame(dist_arr, index = ps_df.index, columns = temp_df.index)
    
    return dist_df
```

```python
def get_knn(filt_distance_df, meta_df):
    res_df = pd.DataFrame(columns = ['closest_sample', 'distance'], index = meta_df.index)
    for col in filt_distance_df.columns:
        min_distance = filt_distance_df[col].min()
        min_indx = filt_distance_df[col].idxmin()
        closest_sample = min_indx
                
        ## insert results to res_df
        res_df.loc[col, 'distance'] = min_distance
        res_df.loc[col, 'closest_sample'] = closest_sample
        
    return res_df
```

#### Find BC nearest neighbor

```python
def get_pcoa(df, metric):
    arr = df.to_numpy()
    bc_df = beta_diversity(counts = arr, ids = list(df.index), metric = metric, validate = False)

    return bc_df.to_data_frame()
```

```python
def get_filt_distance_df(distance_df, ps_meta_df, meta_df):
    cols_to_keep = meta_df.index
    rows_to_keep = ps_meta_df.index
    
    filt_distance_df = distance_df[cols_to_keep]
    filt_distance_df = filt_distance_df[filt_distance_df.index.isin(rows_to_keep)]
    
    return filt_distance_df
```

```python
def get_knn(filt_distance_df, meta_df):
    res_df = pd.DataFrame(columns = ['closest_sample', 'distance'], index = meta_df.index)
    for col in filt_distance_df.columns:
        min_distance = filt_distance_df[col].min()
        min_indx = filt_distance_df[col].idxmin()
        closest_sample = min_indx
                
        ## insert results to res_df
        res_df.loc[col, 'distance'] = min_distance
        res_df.loc[col, 'closest_sample'] = closest_sample
        
    return res_df
```

#### Create DF for plots

```python
def get_meta_cols(df, meta_df, col):
    if not set(col).issubset(set(df.columns)):
        merged = df.merge(meta_df[col], left_index = True, right_index = True, how = 'inner')
        return merged
    else:
        return df
```

```python
both_pca_df = pd.merge(both_pca_df, ps_meta_df[['mt_pseudotime']], how = 'left', on = 'sampleID')
both_pca_df = get_meta_cols(both_pca_df, both_meta_df, ['community', 'DB_type', 'subjectID', 'time'])
```

```python
def order_index(knn_res_df, both_pca_df):
    cros_index = [i for i in range(1, len(knn_res_df) * 2, 2)]
    temp_index =  [i for i in range(0, len(knn_res_df) * 2, 2)]
    
    knn_cros_df = knn_res_df[['closest_sample']].reset_index(level = 0, inplace = False).drop('sampleID', 1)
    knn_cros_df.index = cros_index
    knn_cros_df.rename(columns = {'closest_sample': 'sampleID'}, inplace = True)

    knn_temp_df = knn_res_df.reset_index(level = 0, inplace = False).drop(['closest_sample', 'distance'], 1)
    knn_temp_df.index = temp_index

    conc_knn = pd.concat([knn_temp_df, knn_cros_df], axis = 0, ignore_index = False)
    conc_knn.sort_index(axis = 0, inplace = True)
    conc_knn['sort_index'] = conc_knn.index
    conc_knn.index = conc_knn['sampleID']
    
    ord_pca_df = both_pca_df.reindex(conc_knn.index)
    
    return ord_pca_df
```

```python
def get_df_for_sibject_plot(knn_res_df, both_pca_df, curr_id):
    ## PCA df for curr_id subject
    subject_knn_df = knn_res_df[temp_meta_df['subjectID'] == curr_id]
    subject_pca_df = both_pca_df[both_pca_df.index.isin(subject_knn_df.closest_sample) | both_pca_df.index.isin(subject_knn_df.index)]
    ord_subject_df = order_index(subject_knn_df, subject_pca_df)

    # Add np.nan row every two rows
    s = pd.Series(np.nan, ord_subject_df.columns)
    f = lambda d: d.append(s, ignore_index = True)
    grp = np.arange(len(ord_subject_df)) // 2
    ord_subject_df = ord_subject_df.groupby(grp, group_keys = False).apply(f).reset_index(drop = True)

    ## PCA df for all cros samples that are not knn
    ord_ps_pca_df = pd.merge(ps_pca_df[['PCA1', 'PCA2', 'PCA3']], ps_meta_df[['mt_pseudotime', 'community']], how = 'left', on = 'sampleID')
    subject_cros_lst = subject_knn_df['closest_sample'].tolist()
    all_cros_lst = ord_ps_pca_df.index.tolist()
    not_subject_cros_lst = list(set(all_cros_lst) - set(subject_cros_lst))
    ord_ps_pca_df = ord_ps_pca_df.loc[not_subject_cros_lst, :]

    return ord_subject_df, ord_ps_pca_df
```

```python
## 3d figure with plotly
def get_subject_plotly(ord_ps_pca_df, ord_subject_df, curr_id):
    fig = go.Figure()
    fig.add_trace(go.Scatter3d(x = ord_ps_pca_df['PCA1'],
                     y = ord_ps_pca_df['PCA2'],
                     z = ord_ps_pca_df['PCA3'],
                     mode = 'markers', opacity = 0.7, name = 'Cross-sectional- no neighbors', showlegend = False,
                     marker = dict(color = ord_ps_pca_df['mt_pseudotime'],
                                  line = dict(width = 3,  color = 'DarkSlateGrey'))))
    for i in range(len(ord_subject_df)):
        fig.add_traces(go.Scatter3d(x = ord_subject_df['PCA1'],
                                  y = ord_subject_df['PCA2'],
                                  z = ord_subject_df['PCA3'], 
                                  name = '', showlegend = False, opacity = 1,
                                  mode = 'lines + markers', line = dict(width = 5, color = 'DarkSlateGrey'),
                                  marker = dict(color = ord_subject_df['mt_pseudotime'],
                                               line = dict(width = 3,  color = 'DarkSlateGrey'))))

    fig.update_traces(connectgaps = False)


    fig.update_layout(scene = dict(
                        xaxis_title = 'PCA1',
                        yaxis_title = 'PCA2',
                        zaxis_title = 'PCA3'),
                        title_text = 'Subject number ' + str(curr_id), title_x = 0.5,
                        width = 700)#,
    #                     margin = dict(r = 20, b = 10, l = 10, t = 10))
    fig.show()
    
    return fig
```

```python
def get_all_for_subject_plot(knn_res_df, both_pca_df, curr_id, path):
    ord_subject_df, ord_ps_pca_df = get_df_for_sibject_plot(knn_res_df, both_pca_df, curr_id)
    fig = get_subject_plotly(ord_ps_pca_df, ord_subject_df, curr_id)
    
    fig.write_html(path + 'subject_' + str(curr_id) + '.html')
    
    return fig
```

```python
def get_subject_plots(df):
    x_col_name = df.columns[0]
    y_col_name = df.columns[1]
    z_col_name = df.columns[2]
    x = np.array([df.loc[df['DB_type'] == 'cros', x_col_name], 
                         df.loc[df['DB_type'] == 'temp', x_col_name]])
    y = np.array([df.loc[df['DB_type'] == 'cros', y_col_name], 
                         df.loc[df['DB_type'] == 'temp', y_col_name]])
    z = np.array([df.loc[df['DB_type'] == 'cros', z_col_name], 
                         df.loc[df['DB_type'] == 'temp', z_col_name]])

    ## Variables for 3d plots
    figsize_x = 25
    figsize_y = 10
    points_size = 60
    label_size = 17
    ticks_size = 12
    leg_size = 13
    
    ## Plot 1
    fig1 = plt.figure(figsize = (figsize_x, figsize_y))
    ax1 = plt.axes(projection = '3d')

    for i in range(0, len(x[0])):
        ax1.plot([x[0, i],x[1, i]], [y[0, i],y[1, i]], [z[0, i],z[1, i]], color = 'black', zorder = -1, linewidth = 0.5)
    ax1.scatter(df.loc[df['DB_type'] == 'temp', x_col_name], 
               df.loc[df['DB_type'] == 'temp', y_col_name], 
               df.loc[df['DB_type'] == 'temp', z_col_name], c = 'firebrick', marker = 'o', s = points_size, edgecolor = 'dimgray',
               label = 'temporal')
    ax1.scatter(df.loc[df['DB_type'] == 'cros', x_col_name], 
               df.loc[df['DB_type'] == 'cros', y_col_name], 
               df.loc[df['DB_type'] == 'cros', z_col_name], c = 'teal', marker = 's', s = points_size, edgecolor = 'dimgray',
               label = 'cross-sectional')

    ax1.legend(fontsize = leg_size)
    ax1.set_xlabel(x_col_name, size = label_size, rotation = -22)
    ax1.set_ylabel(y_col_name, size = label_size, rotation = 45)
    ax1.set_zlabel(z_col_name, size = label_size, rotation = 90)
    ax1.tick_params(axis = 'both', which = 'major', labelsize = ticks_size)
    
    
    return fig1, fig2, fig3
```

```python
def get_plots(df):
    x_col_name = df.columns[0]
    y_col_name = df.columns[1]
    z_col_name = df.columns[2]
    x = np.array([df.loc[df['DB_type'] == 'cros', x_col_name], 
                         df.loc[df['DB_type'] == 'temp', x_col_name]])
    y = np.array([df.loc[df['DB_type'] == 'cros', y_col_name], 
                         df.loc[df['DB_type'] == 'temp', y_col_name]])
    z = np.array([df.loc[df['DB_type'] == 'cros', z_col_name], 
                         df.loc[df['DB_type'] == 'temp', z_col_name]])

    ## Variables for 3d plots
    figsize_x = 25
    figsize_y = 10
    points_size = 60
    label_size = 17
    ticks_size = 12
    leg_size = 13
    
    ## Plot 1
    fig1 = plt.figure(figsize = (figsize_x, figsize_y))
    ax1 = plt.axes(projection = '3d')

    for i in range(0, len(x[0])):
        ax1.plot([x[0, i],x[1, i]], [y[0, i],y[1, i]], [z[0, i],z[1, i]], color = 'black', zorder = -1, linewidth = 0.5)
    ax1.scatter(df.loc[df['DB_type'] == 'temp', x_col_name], 
               df.loc[df['DB_type'] == 'temp', y_col_name], 
               df.loc[df['DB_type'] == 'temp', z_col_name], c = 'firebrick', marker = 'o', s = points_size, edgecolor = 'dimgray',
               label = 'temporal')
    ax1.scatter(df.loc[df['DB_type'] == 'cros', x_col_name], 
               df.loc[df['DB_type'] == 'cros', y_col_name], 
               df.loc[df['DB_type'] == 'cros', z_col_name], c = 'teal', marker = 's', s = points_size, edgecolor = 'dimgray',
               label = 'cross-sectional')

    ax1.legend(fontsize = leg_size)
    ax1.set_xlabel(x_col_name, size = label_size, rotation = -22)
    ax1.set_ylabel(y_col_name, size = label_size, rotation = 45)
    ax1.set_zlabel(z_col_name, size = label_size, rotation = 90)
    ax1.tick_params(axis = 'both', which = 'major', labelsize = ticks_size)
    
    ## Plot2
    fig2 = plt.figure(figsize = (figsize_x, figsize_y))
    ax2 = plt.axes(projection = '3d')

    community_colors = {'I':'darkred', 'II':'orange', 'III':'chocolate', 'V':'yellow', 
              'IV':'mediumseagreen', 'IV-A':'mediumpurple', 'IV-B':'royalblue'}
    cmap = plt.cm.rainbow

    for i in range(0, len(x[0])):
        ax2.plot([x[0, i],x[1, i]], [y[0, i],y[1, i]], [z[0, i],z[1, i]], color = 'black', zorder = -1, linewidth = 0.5)
    sc2_1 = ax2.scatter(df.loc[df['DB_type'] == 'temp', x_col_name], 
               df.loc[df['DB_type'] == 'temp', y_col_name], 
               df.loc[df['DB_type'] == 'temp', z_col_name], c = df.loc[df['DB_type'] == 'temp', 'community'].map(community_colors), 
               marker = 'o', s = points_size, edgecolor = 'dimgray', label = 'temporal')
    sc2_2 = ax2.scatter(df.loc[df['DB_type'] == 'cros', x_col_name], 
               df.loc[df['DB_type'] == 'cros', y_col_name], 
               df.loc[df['DB_type'] == 'cros', z_col_name], c = df.loc[df['DB_type'] == 'cros', 'community'].map(community_colors), 
               marker = 's', s = points_size, edgecolor = 'dimgray', label = 'cros-sectional')

    handles = [Line2D([0], [0], marker = 'o', color = 'w', markerfacecolor = v, label = k, markersize = leg_size) 
               for k, v in community_colors.items()]
    ax2.legend(title = 'Community state type', handles = handles, bbox_to_anchor = (1.05, 1), loc = 'upper left')
    ax2.set_xlabel(x_col_name, size = label_size, rotation = -22)
    ax2.set_ylabel(y_col_name, size = label_size, rotation = 45)
    ax2.set_zlabel(z_col_name, size = label_size, rotation = 90)
    ax2.tick_params(axis = 'both', which = 'major', labelsize = ticks_size)
    
    ## Plot 3
    fig3 = plt.figure(figsize = (figsize_x, figsize_y))
    ax3 = plt.axes(projection = '3d')

    for i in range(0, len(x[0])):
        ax3.plot([x[0, i],x[1, i]], [y[0, i],y[1, i]], [z[0, i],z[1, i]], color = 'black', zorder = -1, linewidth = 0.5)
    sc3_1 = ax3.scatter(df.loc[df['DB_type'] == 'temp', x_col_name], 
               df.loc[df['DB_type'] == 'temp', y_col_name], 
               df.loc[df['DB_type'] == 'temp', z_col_name], c = 'grey', 
               marker = 'o', s = points_size, edgecolor = 'dimgray', label = 'temporal')
    sc3_2 = ax3.scatter(df.loc[df['DB_type'] == 'cros', x_col_name], 
               df.loc[df['DB_type'] == 'cros', y_col_name], 
               df.loc[df['DB_type'] == 'cros', z_col_name], c = df.loc[df['DB_type'] == 'cros', 'mt_pseudotime'], 
               marker = 's', s = points_size, edgecolor = 'dimgray', label = 'cross-sectional')

    cbar3 = fig3.colorbar(sc3_1)
    cbar3.set_label("Pseudotime", labelpad =+ 0.5, fontsize = leg_size)
    ax3.legend(fontsize = leg_size)
    ax3.set_xlabel(x_col_name, size = label_size, rotation = -22)
    ax3.set_ylabel(y_col_name, size = label_size, rotation = 45)
    ax3.set_zlabel(z_col_name, size = label_size, rotation = 90)
    ax3.tick_params(axis = 'both', which = 'major', labelsize = ticks_size)
    
    return fig1, fig2, fig3
```

```python
euc_fig1, euc_fig2, euc_fig3 = get_plots(euc_plot_pca_df)
```
