```python
def get_simpson(df):
    ''' add column named simpson to df, 
    '''
    sub_groups = df.groupby(["subjectID", "CST"])['sampleID'].count().to_frame().rename(columns = {'sampleID': 'count'})
    sub_groups.reset_index('CST', drop = False, inplace = True)
    sub_groups = sub_groups.merge(sub_groups.groupby('subjectID').sum().rename(columns = {'count':'total'}), on = 'subjectID')
    
    sub_groups['n'] = (sub_groups['count'] / sub_groups['total'])
    sub_groups['n_divided_N'] = sub_groups['n'] * sub_groups['n']
    sub_res_n = sub_groups.groupby('subjectID')['n_divided_N'].sum().to_frame()
    sub_res_n['simpson'] = 1 - sub_res_n['n_divided_N']
    
    df = df.merge(sub_res_n, how = 'left', on = 'subjectID')
    
    return df, sub_res_n
```

```python
df, sub_res_n = get_simpson(temp_meta)
```

```python
def get_continuous_clinic_col(df, orig_df, group_col, col, first = False):
    grouped_df = orig_df.groupby(group_col).apply(lambda s: pd.Series({ 
        col + '_num_samples': s[col].count(), 
        col + '_mean': s[col].mean(skipna = True), 
        col + '_min': s[col].min(skipna = True), 
        col + '_max': s[col].max(skipna = True)}))
    if first:
        return grouped_df
    else:
        return df.join(grouped_df)
    

def get_cat_clinic_col(df, orig_df, group_col, col):
    grouped_df = orig_df.groupby(group_col).apply(lambda s: pd.Series({ 
        col + '_num_samples': s[col].count(), 
        col + '_positive': s[col].sum(skipna = True), 
        col + '_negative': s[col].eq(0).sum()}))
    df = df.join(grouped_df)
    
    return df
```

```python
clin_k_meta = k_sub_temp_meta.set_index('subjectID').copy()
clin_k_meta = get_continuous_clinic_col(clin_k_meta, temp_meta, 'subjectID', 'nugent')
clin_k_meta = get_continuous_clinic_col(clin_k_meta, temp_meta, 'subjectID', 'ph')
clin_k_meta = get_continuous_clinic_col(clin_k_meta, temp_meta, 'subjectID', 'age')
clin_k_meta = get_continuous_clinic_col(clin_k_meta, temp_meta, 'subjectID', 'shannon_index')
clin_k_meta = get_cat_clinic_col(clin_k_meta, temp_meta, 'subjectID', 'ethnicity')
clin_k_meta = get_cat_clinic_col(clin_k_meta, temp_meta, 'subjectID', 'BV')
```

#### Krskal wallis

```python
def get_kruskal(df, y):
    tmp = df[['k', y]].dropna(subset = ['k', y], axis = 0)
    
    group_lst = []
    for k_num in tmp.k.unique():
        group = tmp[y][tmp['k'] == k_num].unique()
        group_lst.append(group)

    pval = stats.kruskal(*group_lst).pvalue
    df_dun = sp.posthoc_dunn(a = group_lst, p_adjust  = 'fdr_bh')
    
    print(y, 'p. value: ', pval)
    if pval <= 0.05:
        display(df_dun)
    
    return pval, df_dun
```
